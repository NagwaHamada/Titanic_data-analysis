```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Soft\Titanic.csv")
```

```python
Dnull = df.isnull().sum()
Dnull
```

```python
Dnull[Dnull>0].plot(kind='bar')
```

```python
sex_count = df['Sex'].value_counts()
sex_count
plt.bar(sex_count.index, sex_count.values, color='purple',  width=0.4)
plt.show()
```

```python
df['Age'].hist(color='gray')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution of Titanic Passengers')


```

```python
df[df['Survived']==0]['Age'].hist()
```

```python
sex_count.plot(kind='pie',autopct = "%0.2f%%")
```

```python
counts = pd.crosstab(df['Sex'], df['Survived'])

counts.plot(kind='bar', stacked=False, color=['purple','gray'])
plt.xlabel("Sex")
plt.ylabel("Count")
plt.title("Survival Count by Gender")
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python
count = pd.crosstab(df["Pclass"], df["Survived"])

count.plot(kind="bar", color=["purple", "gray"])
plt.title("Survival  Based on Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.legend(["Not Survived", "Survived"])
plt.show()
```

```python

df["FamilySize"] = df["SibSp"] + df["Parch"]

count = pd.crosstab(df["FamilySize"], df["Survived"])
count
```

```python
count.plot(kind="bar", color=["gray", "purple"])

plt.title("Survival Count by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Count")
plt.legend(["Not Survived", "Survived"])

plt.show()
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"]

family_survival_rate = df.groupby("FamilySize")["Survived"].mean()

plt.plot(family_survival_rate.index, family_survival_rate.values, marker="o", color="purple",mec="grey",mfc="grey",ms=10)

plt.title("Survival Rate by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Survival Rate")
plt.grid(True)


plt.show()
```
